En stang AB med masse $m$ er leddlagret i punkt A og understttet av en kloss i punkt D. Klossen kan bevege seg friksjonsfritt i horisontalplanet. En horisontal kraft $P$ virker p klossen, se figuren. Det er ingen friksjon i D. Resultanten av egenvekten $mg$ til stanga angriper i stangens tyngdepunkt.

Oppgaven skal utfres i to steg. Frst skal uttrykket for kraften $P$ evalueres og plottes for $x$ mellom 50 og 750 mm. S skal den strste verdien av $P$, og ved hvilken avstand $x$ den opptrer ved, finnes.

Frst er vi ndt  importere funksjonsbibliotek for matematiske beregninger og for plotting.

````python
import matplotlib.pyplot as plt
import numpy as np
````

Vi definerer noen konstanter som vi kommer til  trenge.

````python
L = 1000.0 / 1000.0   # Lengde av stav [m]
h = 400.0  / 1000.0   # Hyde av kloss [m]
m = 10.0              # Stangas masse [kg]
g = 9.81              # Gravitasjonskonstant [m/s^2]
````

Frste steg til  plotte funksjonen for kraften,

\begin{equation*}
P = \frac{L_{AB} \cdot h \cdot mg}{2(x^2+h^2)^{3/2}} \cdot x.
\end{equation*}

er  evaluere uttrykket for et sett verdier av $x$ mellom 50 og 750mm. 

Det kan vi enkelt gjre gjennom numpy-funksjonen  
```python 
np.arange(Start, Stop, Steglengde)
```  
Denne ligner p  
```python 
np.linspace()
```  
som ble brukt i Pythonving 1, men lager et array med fast steglengde istedet for et med gitt antall elementer. 

````python
x = np.arange(50.0, 750.0, 1.0)/1000.0 # m
````

****

<div class="alert alert-success">

<p> Regn ut kraften $P$ for alle verdier av $x$ (50-750mm), og plott $P$ mot $x$. Husk  skrive ut verdiene ($\textit{print("svar")}$) og navngi aksene i plottet.

Bruk pflgende kodecelle.
</p>
</div>

````python
def P(x):
    kraft=x*((L*h*m*g)/(2*(x**2+h**2)**(3/2)))
    return kraft

plt.plot(x, P(x), label="P(x) for L={:.2f}m, h={:.2f}m, m={:.1f}kg".format(L, h, m))
plt.xlabel("x [mm]")
plt.ylabel("P [N]")
plt.legend()
plt.show()

print(f"Verdiene er: {P(x)}")
````

<div class="alert alert-success">

Finn maksimal kraft $P_{max}$ og ved hvilken avstand, $x_{P,max}$, den vil opptre ved. Se hintet nedenfor for nyttige funksjoner for  finne maksimal verdi i et array. 

Plot grafen ovenfor p nytt (i kodecellen under) og inkluder punktet ved maks kraft ($x_{P,max}$, $P_{max}$).

</div>

````python
Parray=np.array(P(x))
P_max=np.max(Parray)
P_max_index=np.argmax(Parray)
x_pmax=x[P_max_index]

plt.plot(x, P(x), label="P(x) for L={:.2f}m, h={:.2f}m, m={:.1f}kg".format(L, h, m))
plt.scatter(x_pmax, P_max, label="(x_pmax, P_max)")
plt.xlabel("x [mm]")
plt.ylabel("P [N]")
plt.legend()
plt.show()
````
