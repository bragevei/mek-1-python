En uendelig stiv, dvs. ikke-deformerbar, bjelke AB er hengt opp i to deformerbare skråstag AC og BD, se figuren rett under. Konstruksjonens belastning og geometri er symmetrisk. Punktene A og B beveger seg derfor vertikalt (rett nedover) når stagene AC og BD forlenges, og bjelken AB forholder seg horisontal. Stagene har lengde $L_0 = 5,2$ m når bjelken er ubelastet. Ved økende belastning $q$ får punkt A (og B) en vertikalforskyvning $v$, og stagene strekkes til lengde $L$.

****

I denne pythonoppgaven skal det lages to plott. I det første skal dere vise og sammenligne stagets forlengelse $\Delta L$ som funksjon av bjelkens forskyvning $\nu$. Det skal gjøres for både en eksakt og en lineariserte modell. Videre skal dere regne ut og plotte prosentvis avvik mellom de to modellene. 

Forlengelsen av stagene er gitt som $\Delta L = L -L_0$. Initiell lengde $L_0=5,2$ m jfr. figuren ovenfor og $v$ skal gå fra 0 til 2 m.  

Stagets lengde etter den eksakte modellen kan finnes som, 

$L = L_0 \sqrt{1+\frac{24}{13} \frac{\nu}{L_0}+\Big(\frac{\nu}{L_0}\Big)^2}$, 

og etter den lineariserte modellen som,

$L = L_0 \Big(1 + \frac{12}{13} \frac{\nu}{L_0} \Big)$.

# Funksjonsbilbiotek og definisjon av parametre

Vi kommer til å trenge ulike matematiske funksjoner og plotteverktøy, og trenger derfor å importere funksjonsbibliotekene NumPy og MatPlotLib.pyplot. I denne øvingen er det gjort for dere, men senere er dere nødt å huske på det selv. Disse forhåndsbestemte cellene blir kjørt om du trykker "Run All". 

````python
import numpy as np
import matplotlib.pyplot as plt
````

Så definerer vi initiell staglengde $L_0$, og et array (NumPy-liste) med vertikalforskyvninger $v$. Ettersom vi ønsker å evaluere forlengelsen av staget ved mange ulike forskyvninger, definerer vi dem i nettopp et array ved hjelp av funksjonen `np.linspace(start, stop, ant. elementer)`.

````python
L0 = 5.2 # [m]
vertikalForskyvning = np.linspace(0, 2, 100) #  [m], En numpy-liste med 50 elementer som går fra 0 til 2
````

# Beregning av stagets forlengelse

<div class="alert alert-success">

Regn ut stagets forlengelse, $\Delta L$, fra den eksakte og den lineariserte formuleringen. 

Svaret skal plottes i neste deloppgave.

Husk å trykke ">|Run" fra verktøylinjen eller "shift"+"enter" på tastaturet på hver celle for å faktisk kjøre koden.

Bruk kodecellen under.

***
</div>

````python
def delta_L(L0, vertikalForskyvning):
    L_eksakt=L0*np.sqrt(1+24/13*vertikalForskyvning/L0+(vertikalForskyvning/L0)**2)
    L_lin=L0*(1+(12*vertikalForskyvning)/(13*L0))
    return L_eksakt-L0, L_lin-L0

delta_L(L0, vertikalForskyvning)
````

****

# Plot av stagets forlengelse

<div class="alert alert-success">

Plot resultatene ved hjelp av funksjonene fra MatPlotLib.pyplot som ovenfor ble forkortet `plt`. Husk å navngi aksene, definere en legend (forklaring på hva som er eksakt og linearisert modell) for plottene og lage tittel. 

Bruk kodecellen under.

</div>

````python

plt.plot(delta_L(L0, vertikalForskyvning)[0], vertikalForskyvning, label="Eksakt modell")
plt.plot(delta_L(L0, vertikalForskyvning)[1], vertikalForskyvning, label="Linearisert modell")
plt.xlabel("Forskyvning i vertikal retning")
plt.ylabel("Endring i stagets lengde")
plt.legend()
plt.show()
````

# Avvik mellom eksakt og linearisert modell

<div class="alert alert-success">

Vi ønsker nå å undersøke forskjellen mellom de to modellene. Beregn og plot derfor prosentvis avvik fra eksakt modell for alle vertikalforskyvningene brukt ovenfor. 


Les av fra grafen hvor stor $v$ kan være før den gir et avvik større enn 1% og sammenlign med løsningsantydningen under "Show Solution".

Husk at $\Delta L(0)=0$ for begge formuleringer. Du må derfor ekskludere første element i $\Delta L$ for å regne ut prosentvis endring da dette innebære å dele på 0.  

Benytt påfølgende kodecelle for beregning og plotting.

***

</div>

````python
def avvik(eksakt_mod, lin_mod, vertikalForskyvning):
    prosent=np.zeros_like(eksakt_mod)
    v=None
    for i in range(1, len(eksakt_mod)):
        prosent[i]=(abs(eksakt_mod[i]-lin_mod[i])/eksakt_mod[i])*100
        if v==None and prosent[i] >= 1:
            v=vertikalForskyvning[i]
    return v, prosent

eksakt_mod, lin_mod = delta_L(L0, vertikalForskyvning)
v, p = avvik(eksakt_mod, lin_mod, vertikalForskyvning)
print(f"For v>={v:.3f} har vi et avvik større enn 1%")

plt.plot(vertikalForskyvning, p, label="Prosentvis avvik fra eksakt modell")
plt.xlabel("Avvik fra eksakt modell")
plt.ylabel("Prosentvis endring")
plt.legend()
plt.show()
````
